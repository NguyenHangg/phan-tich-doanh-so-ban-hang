```python
# Cell 1: Import thư viện
print("="*80)
print("PHÂN LỚP - DỰ ĐOÁN PHÂN KHÚC KHÁCH HÀNG")
print("="*80)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                           roc_auc_score, f1_score, accuracy_score)
from sklearn.metrics import precision_recall_curve, roc_curve, auc
import warnings
from pathlib import Path
import joblib

# Kiểm tra phiên bản scikit-learn
import sklearn
print(f"📌 Scikit-learn version: {sklearn.__version__}")

warnings.filterwarnings('ignore')

# Thiết lập hiển thị
plt.rcParams['figure.figsize'] = (14, 8)
sns.set_style('whitegrid')
pd.set_option('display.max_columns', None)

print("✅ Đã import thư viện thành công!")

# Cell 2: Đường dẫn và đọc dữ liệu
print("\n" + "="*80)
print("ĐỌC DỮ LIỆU")
print("="*80)

PROJECT_DIR = Path.cwd() / 'vietnam_retail_analysis'
DATA_DIR = PROJECT_DIR / 'data' / 'processed'
OUTPUTS_DIR = PROJECT_DIR / 'outputs'
MODELS_DIR = OUTPUTS_DIR / 'models'

MODELS_DIR.mkdir(parents=True, exist_ok=True)

print(f"📁 Thư mục dữ liệu: {DATA_DIR}")
print(f"📁 Thư mục kết quả: {OUTPUTS_DIR}")
print(f"📁 Thư mục models: {MODELS_DIR}")

# Đọc dữ liệu gốc
sales_file = DATA_DIR / 'combined_sales_data.csv'
if not sales_file.exists():
    sales_file = PROJECT_DIR / 'data' / 'raw' / 'combined_sales_data.csv'

if sales_file.exists():
    df_sales = pd.read_csv(sales_file, encoding='utf-8-sig')
    print(f"✅ Đã đọc file bán hàng: {sales_file}")
else:
    print("⚠️ Không tìm thấy file bán hàng, tạo dữ liệu mẫu...")
    np.random.seed(42)
    n = 5000
    df_sales = pd.DataFrame({
        'Mã khách hàng': [f'KH-{i:05d}' for i in range(1, n+1)],
        'Doanh số': np.random.randint(100000, 5000000, n),
        'Số lượng': np.random.randint(1, 20, n)
    })

# Đọc kết quả phân cụm
segments_file = OUTPUTS_DIR / 'customer_segments.csv'
if segments_file.exists():
    df_segments = pd.read_csv(segments_file, encoding='utf-8-sig')
    print(f"✅ Đã đọc file phân cụm: {segments_file}")
else:
    print("⚠️ Không tìm thấy file phân cụm, tạo dữ liệu mẫu...")
    n = 500
    df_segments = pd.DataFrame({
        'Mã khách hàng': [f'KH-{i:05d}' for i in range(1, n+1)],
        'Cluster': np.random.randint(0, 4, n),
        'Cluster_Name': [f'Cụm {np.random.randint(0, 4)}' for _ in range(n)]
    })

print(f"✅ Dữ liệu bán hàng: {df_sales.shape}")
print(f"✅ Dữ liệu phân cụm: {df_segments.shape}")

# Cell 3: Xác định các cột
print("\n" + "="*80)
print("XÁC ĐỊNH CÁC CỘT DỮ LIỆU")
print("="*80)

# Xác định cột mã khách hàng
if 'Mã khách hàng' in df_sales.columns:
    customer_col = 'Mã khách hàng'
elif 'Customer ID' in df_sales.columns:
    customer_col = 'Customer ID'
else:
    # Tìm cột có chứa 'khách' hoặc 'customer'
    customer_cols = [col for col in df_sales.columns if 'khách' in col.lower() or 'customer' in col.lower()]
    if customer_cols:
        customer_col = customer_cols[0]
    else:
        customer_col = df_sales.columns[0]
        print(f"⚠️ Sử dụng cột {customer_col} làm mã khách hàng")

# Xác định cột doanh số
if 'Doanh số' in df_sales.columns:
    sales_col = 'Doanh số'
elif 'Sales' in df_sales.columns:
    sales_col = 'Sales'
else:
    sales_cols = [col for col in df_sales.columns if 'doanh' in col.lower() or 'sales' in col.lower() or 'revenue' in col.lower()]
    if sales_cols:
        sales_col = sales_cols[0]
    else:
        numeric_cols = df_sales.select_dtypes(include=[np.number]).columns.tolist()
        sales_col = numeric_cols[0] if numeric_cols else df_sales.columns[1]
        print(f"⚠️ Sử dụng cột {sales_col} làm doanh số")

print(f"\n✅ Các cột được xác định:")
print(f"   - Khách hàng: {customer_col}")
print(f"   - Doanh số: {sales_col}")

# Cell 4: Tổng hợp đặc trưng khách hàng
print("\n" + "="*80)
print("TỔNG HỢP ĐẶC TRƯNG KHÁCH HÀNG")
print("="*80)

# Xây dựng dictionary aggregation linh hoạt
agg_dict = {
    sales_col: ['sum', 'mean', 'std', 'count']
}

# Thêm cột số lượng nếu có
if 'Số lượng' in df_sales.columns:
    agg_dict['Số lượng'] = ['sum', 'mean']
elif 'Quantity' in df_sales.columns:
    agg_dict['Quantity'] = ['sum', 'mean']

# Thêm các cột số khác có thể hữu ích
numeric_cols = df_sales.select_dtypes(include=[np.number]).columns.tolist()
other_cols = [col for col in numeric_cols if col not in [sales_col, 'Số lượng', 'Quantity']]
for col in other_cols[:3]:  # Chỉ lấy tối đa 3 cột số khác
    agg_dict[col] = ['mean', 'std']

print(f"📊 Các đặc trưng sẽ được tổng hợp:")
for col, aggs in agg_dict.items():
    print(f"   - {col}: {aggs}")

# Thực hiện tổng hợp
customer_features = df_sales.groupby(customer_col).agg(agg_dict).round(2)

# Đặt tên cột
customer_features.columns = ['_'.join(col).strip() for col in customer_features.columns.values]
customer_features = customer_features.reset_index()
customer_features.columns = [customer_col] + [col for col in customer_features.columns if col != customer_col]

print(f"\n✅ Đã tạo {len(customer_features.columns)-1} features cho {len(customer_features)} khách hàng")
print("\n📊 Features:")
print(customer_features.columns.tolist())

# Cell 5: Kết hợp với nhãn phân cụm
print("\n" + "="*80)
print("KẾT HỢP VỚI NHÃN PHÂN CỤM")
print("="*80)

# Chuẩn bị dữ liệu phân cụm
if 'Mã khách hàng' in df_segments.columns:
    seg_customer_col = 'Mã khách hàng'
elif 'Customer ID' in df_segments.columns:
    seg_customer_col = 'Customer ID'
else:
    seg_customer_col = df_segments.columns[0]

# Lấy cột cluster
cluster_col = None
for col in ['Cluster', 'cluster', 'Cụm', 'cụm']:
    if col in df_segments.columns:
        cluster_col = col
        break

if cluster_col is None:
    cluster_col = df_segments.columns[1]  # Giả sử cột thứ 2 là cluster

df_segments_clean = df_segments[[seg_customer_col, cluster_col]].drop_duplicates()

# Kết hợp dữ liệu
data = customer_features.merge(df_segments_clean, left_on=customer_col, right_on=seg_customer_col, how='inner')

print(f"✅ Dữ liệu sau khi kết hợp: {data.shape}")
print("\n📊 5 dòng đầu:")
print(data.head())

# Cell 6: Phân tích phân bố nhãn
print("\n" + "="*80)
print("PHÂN BỐ NHÃN PHÂN LỚP")
print("="*80)

# Phân bố cluster
cluster_dist = data[cluster_col].value_counts().sort_index()
print("📊 Phân bố các cụm:")
for cluster, count in cluster_dist.items():
    print(f"  - Cụm {cluster}: {count} khách hàng ({count/len(data)*100:.1f}%)")

# Vẽ biểu đồ
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Pie chart
axes[0].pie(cluster_dist.values, labels=[f'Cụm {i}' for i in cluster_dist.index], 
           autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3(range(len(cluster_dist))))
axes[0].set_title('Phân bố các cụm')

# Bar chart
colors = plt.cm.viridis(np.linspace(0, 1, len(cluster_dist)))
bars = axes[1].bar(range(len(cluster_dist)), cluster_dist.values, color=colors)
axes[1].set_xlabel('Cụm')
axes[1].set_ylabel('Số lượng')
axes[1].set_title('Số lượng theo cụm')
axes[1].set_xticks(range(len(cluster_dist)))
axes[1].set_xticklabels([f'Cụm {i}' for i in cluster_dist.index])

# Thêm giá trị
for bar, count in zip(bars, cluster_dist.values):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height,
                f'{count}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'classification_target.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 7: Chuẩn bị features và target
print("\n" + "="*80)
print("CHUẨN BỊ FEATURES VÀ TARGET")
print("="*80)

# Xác định feature columns
feature_cols = [col for col in data.columns if col not in [customer_col, seg_customer_col, cluster_col]]
X = data[feature_cols]
y = data[cluster_col]

print(f"📊 Features ({len(feature_cols)} biến):")
for i, col in enumerate(feature_cols[:10]):
    print(f"   {i+1}. {col}")
if len(feature_cols) > 10:
    print(f"   ... và {len(feature_cols)-10} features khác")

print(f"\n📊 Target: {y.nunique()} classes")
print(y.value_counts().sort_index())

# Mã hóa nhãn
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("\n✅ Nhãn sau khi mã hóa:")
for i, label in enumerate(le.classes_):
    print(f"  - Cụm {label} → {i}")

# Cell 8: Xử lý missing values và chia train/test
print("\n" + "="*80)
print("XỬ LÝ DỮ LIỆU VÀ CHIA TRAIN/TEST")
print("="*80)

# Kiểm tra missing values
missing = X.isnull().sum()
if missing.sum() > 0:
    print("⚠️ Phát hiện missing values:")
    print(missing[missing > 0])
    X = X.fillna(X.median())
    print("✅ Đã xử lý missing values")

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

print(f"✅ Kích thước tập train: {X_train.shape}")
print(f"✅ Kích thước tập test: {X_test.shape}")

# Chuẩn hóa
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("✅ Đã chuẩn hóa dữ liệu")

# Cell 9: Hàm đánh giá mô hình
print("\n" + "="*80)
print("ĐỊNH NGHĨA HÀM ĐÁNH GIÁ")
print("="*80)

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """
    Đánh giá mô hình phân lớp
    """
    print(f"\n{'='*50}")
    print(f"📊 ĐÁNH GIÁ: {model_name}")
    print(f"{'='*50}")
    
    # Huấn luyện
    model.fit(X_train, y_train)
    
    # Dự đoán
    y_pred = model.predict(X_test)
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1_macro = f1_score(y_test, y_pred, average='macro')
    f1_weighted = f1_score(y_test, y_pred, average='weighted')
    
    print(f"\n🎯 ĐỘ CHÍNH XÁC:")
    print(f"  - Accuracy: {accuracy:.4f}")
    print(f"  - F1-Score (macro): {f1_macro:.4f}")
    print(f"  - F1-Score (weighted): {f1_weighted:.4f}")
    
    # Tính ROC-AUC nếu có probability
    if hasattr(model, "predict_proba"):
        try:
            y_pred_proba = model.predict_proba(X_test)
            if len(le.classes_) == 2:  # Binary classification
                roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
            else:  # Multi-class
                roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr', average='macro')
            print(f"  - ROC-AUC: {roc_auc:.4f}")
        except:
            print("  - ROC-AUC: Không thể tính")
    
    # Classification Report
    print(f"\n📋 CLASSIFICATION REPORT:")
    target_names = [f'Cụm {i}' for i in le.classes_]
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Vẽ confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=target_names, yticklabels=target_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / f'confusion_matrix_{model_name.replace(" ", "_")}.png')
    plt.show()
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1_macro')
    print(f"\n📊 CROSS-VALIDATION (5-fold):")
    print(f"  - Mean F1: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    
    return {
        'model': model,
        'name': model_name,
        'accuracy': accuracy,
        'f1_macro': f1_macro,
        'f1_weighted': f1_weighted,
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'y_pred': y_pred
    }

# Cell 10: Huấn luyện các mô hình
print("\n" + "="*80)
print("HUẤN LUYỆN CÁC MÔ HÌNH")
print("="*80)

results = []

# 1. Logistic Regression - SỬA LỖI Ở ĐÂY
print("\n🔍 ĐANG HUẤN LUYỆN LOGISTIC REGRESSION...")
# Không dùng multi_class='ovr' nếu phiên bản cũ không hỗ trợ
lr = LogisticRegression(random_state=42, max_iter=1000)
# Hoặc dùng: lr = LogisticRegression(random_state=42, max_iter=1000, multi_class='auto')
lr_result = evaluate_model(lr, X_train_scaled, X_test_scaled, y_train, y_test, "Logistic Regression")
results.append(lr_result)

# 2. Decision Tree
print("\n🔍 ĐANG HUẤN LUYỆN DECISION TREE...")
dt = DecisionTreeClassifier(random_state=42, max_depth=5)
dt_result = evaluate_model(dt, X_train_scaled, X_test_scaled, y_train, y_test, "Decision Tree")
results.append(dt_result)

# 3. Random Forest
print("\n🔍 ĐANG HUẤN LUYỆN RANDOM FOREST...")
rf = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=10)
rf_result = evaluate_model(rf, X_train_scaled, X_test_scaled, y_train, y_test, "Random Forest")
results.append(rf_result)

# 4. Gradient Boosting
print("\n🔍 ĐANG HUẤN LUYỆN GRADIENT BOOSTING...")
gb = GradientBoostingClassifier(random_state=42, n_estimators=100, max_depth=5)
gb_result = evaluate_model(gb, X_train_scaled, X_test_scaled, y_train, y_test, "Gradient Boosting")
results.append(gb_result)

# Cell 11: So sánh các mô hình
print("\n" + "="*80)
print("SO SÁNH CÁC MÔ HÌNH")
print("="*80)

# Tạo bảng so sánh
comparison_df = pd.DataFrame([{
    'Mô hình': r['name'],
    'Accuracy': r['accuracy'],
    'F1-Macro': r['f1_macro'],
    'F1-Weighted': r['f1_weighted'],
    'CV Mean': r['cv_mean'],
    'CV Std': r['cv_std']
} for r in results])

comparison_df = comparison_df.sort_values('F1-Macro', ascending=False)
print("\n📊 BẢNG SO SÁNH:")
print(comparison_df.to_string(index=False))

# Vẽ biểu đồ so sánh
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# So sánh F1
models = comparison_df['Mô hình'].values
f1_scores = comparison_df['F1-Macro'].values
colors = ['gold' if x == max(f1_scores) else 'skyblue' for x in f1_scores]
bars1 = axes[0].barh(models, f1_scores, color=colors)
axes[0].set_xlabel('F1-Score (Macro)')
axes[0].set_title('So sánh F1-Score')
axes[0].set_xlim(0, 1)
for bar, score in zip(bars1, f1_scores):
    axes[0].text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2, 
                f'{score:.4f}', va='center')

# So sánh CV
cv_means = comparison_df['CV Mean']
cv_stds = comparison_df['CV Std']
bars2 = axes[1].bar(models, cv_means, yerr=cv_stds*2, capsize=5,
                    color=['gold' if x == max(cv_means) else 'skyblue' for x in cv_means])
axes[1].set_ylabel('CV Score')
axes[1].set_title('Cross-validation Score')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'model_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 12: Chọn mô hình tốt nhất
print("\n" + "="*80)
print("MÔ HÌNH TỐT NHẤT")
print("="*80)

best_idx = comparison_df['F1-Macro'].idxmax()
best_model_name = comparison_df.loc[best_idx, 'Mô hình']
best_model = results[best_idx]['model']

print(f"🏆 Mô hình tốt nhất: {best_model_name}")
print(f"   F1-Macro: {comparison_df.loc[best_idx, 'F1-Macro']:.4f}")
print(f"   Accuracy: {comparison_df.loc[best_idx, 'Accuracy']:.4f}")

# Feature importance nếu có
if hasattr(best_model, 'feature_importances_'):
    importances = best_model.feature_importances_
    feature_importance = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': importances
    }).sort_values('Importance', ascending=False)
    
    print("\n📊 TOP 10 FEATURE IMPORTANCE:")
    print(feature_importance.head(10))
    
    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    top_features = feature_importance.head(10)
    plt.barh(range(len(top_features)), top_features['Importance'], color='coral')
    plt.yticks(range(len(top_features)), top_features['Feature'])
    plt.xlabel('Importance')
    plt.title('Top 10 Feature Importance')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'feature_importance.png', dpi=300, bbox_inches='tight')
    plt.show()

elif hasattr(best_model, 'coef_'):
    # Cho Logistic Regression
    coef = best_model.coef_
    avg_coef = np.mean(np.abs(coef), axis=0)
    feature_importance = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': avg_coef
    }).sort_values('Importance', ascending=False)
    
    print("\n📊 TOP 10 FEATURE IMPORTANCE (based on coefficients):")
    print(feature_importance.head(10))

# Cell 13: Lưu mô hình
print("\n" + "="*80)
print("LƯU MÔ HÌNH")
print("="*80)

# Lưu mô hình
model_path = MODELS_DIR / f'best_model_{best_model_name.replace(" ", "_")}.pkl'
joblib.dump(best_model, model_path)

# Lưu scaler
scaler_path = MODELS_DIR / 'scaler.pkl'
joblib.dump(scaler, scaler_path)

# Lưu label encoder
le_path = MODELS_DIR / 'label_encoder.pkl'
joblib.dump(le, le_path)

# Lưu so sánh
comparison_df.to_csv(OUTPUTS_DIR / 'model_comparison.csv', index=False, encoding='utf-8-sig')

print(f"✅ Đã lưu mô hình: {model_path}")
print(f"✅ Đã lưu scaler: {scaler_path}")
print(f"✅ Đã lưu label encoder: {le_path}")
print(f"✅ Đã lưu so sánh: {OUTPUTS_DIR}/model_comparison.csv")

# Cell 14: Demo dự đoán
print("\n" + "="*80)
print("DEMO DỰ ĐOÁN CHO KHÁCH HÀNG MỚI")
print("="*80)

def predict_customer(features, model, scaler, le):
    """
    Dự đoán phân khúc cho khách hàng mới
    """
    features_scaled = scaler.transform([features])
    pred = model.predict(features_scaled)[0]
    
    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(features_scaled)[0]
        confidence = proba[pred]
    else:
        confidence = 1.0
        proba = None
    
    segment = le.inverse_transform([pred])[0]
    
    return segment, confidence, proba

# Lấy mẫu từ test set
sample_idx = 0
sample_features = X_test.iloc[sample_idx].values
true_segment = le.inverse_transform([y_test[sample_idx]])[0]

# Dự đoán
pred_segment, confidence, proba = predict_customer(
    sample_features, best_model, scaler, le
)

print(f"\n📌 Khách hàng mẫu:")
print(f"   - Phân khúc thực tế: Cụm {true_segment}")
print(f"   - Phân khúc dự đoán: Cụm {pred_segment}")
print(f"   - Độ tin cậy: {confidence:.2%}")

if proba is not None:
    print("\n📊 Xác suất cho từng cụm:")
    for i, (cluster, p) in enumerate(zip(le.classes_, proba)):
        bar = '█' * int(p * 20)
        print(f"   - Cụm {cluster}: {p:.2%} {bar}")

# Cell 15: Kết luận
print("\n" + "="*80)
print("KẾT LUẬN PHÂN LỚP")
print("="*80)

print(f"""
1. TỔNG QUAN:
   - Bài toán: Dự đoán phân khúc khách hàng
   - Số mẫu: {len(X)}
   - Số features: {len(feature_cols)}
   - Số classes: {len(le.classes_)}

2. KẾT QUẢ TỐT NHẤT:
   - Mô hình: {best_model_name}
   - F1-Macro: {comparison_df.loc[best_idx, 'F1-Macro']:.4f}
   - Accuracy: {comparison_df.loc[best_idx, 'Accuracy']:.4f}
   - Cross-validation: {comparison_df.loc[best_idx, 'CV Mean']:.4f} (+/- {comparison_df.loc[best_idx, 'CV Std']*2:.4f})

3. ỨNG DỤNG:
   - Tự động phân loại khách hàng mới
   - Cá nhân hóa chiến lược marketing
   - Tối ưu nguồn lực chăm sóc khách hàng

4. FILE ĐÃ LƯU:
   - model_comparison.csv: So sánh các mô hình
   - best_model_*.pkl: Mô hình tốt nhất
   - scaler.pkl: Bộ chuẩn hóa
   - label_encoder.pkl: Bộ mã hóa nhãn
""")

print("\n" + "="*80)
print("✅ HOÀN THÀNH PHÂN LỚP!")
print("="*80)
```
