```python
# Cell 1: Import thư viện
print("="*80)
print("HỒI QUY VÀ CHUỖI THỜI GIAN - DỰ BÁO DOANH SỐ")
print("="*80)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import warnings
from pathlib import Path
import itertools
from datetime import datetime, timedelta

warnings.filterwarnings('ignore')

# Thiết lập hiển thị
plt.rcParams['figure.figsize'] = (16, 10)
sns.set_style('whitegrid')
pd.set_option('display.max_columns', None)

print("✅ Đã import thư viện thành công!")

# Cell 2: Đường dẫn và đọc dữ liệu
print("\n" + "="*80)
print("ĐỌC DỮ LIỆU")
print("="*80)

PROJECT_DIR = Path.cwd() / 'vietnam_retail_analysis'
DATA_DIR = PROJECT_DIR / 'data' / 'processed'
OUTPUTS_DIR = PROJECT_DIR / 'outputs'

print(f"📁 Thư mục dữ liệu: {DATA_DIR}")
print(f"📁 Thư mục kết quả: {OUTPUTS_DIR}")

# Đọc dữ liệu
sales_file = DATA_DIR / 'combined_sales_data.csv'
if not sales_file.exists():
    sales_file = PROJECT_DIR / 'data' / 'raw' / 'combined_sales_data.csv'

if sales_file.exists():
    df = pd.read_csv(sales_file, encoding='utf-8-sig')
    print(f"✅ Đã đọc file: {sales_file}")
else:
    print("⚠️ Không tìm thấy file, tạo dữ liệu mẫu...")
    np.random.seed(42)
    dates = pd.date_range('2022-01-01', '2023-12-31', freq='D')
    n = len(dates)
    
    # Tạo xu hướng
    trend = np.linspace(1000000, 2000000, n)
    # Tạo mùa vụ
    seasonal = 500000 * np.sin(2 * np.pi * np.arange(n) / 365)
    # Nhiễu
    noise = np.random.normal(0, 200000, n)
    
    sales = trend + seasonal + noise
    sales = np.maximum(sales, 0)  # Không âm
    
    df = pd.DataFrame({
        'Ngày đặt': dates,
        'Doanh số': sales.astype(int)
    })
    print(f"✅ Đã tạo dữ liệu mẫu: {len(df)} ngày")

# Xử lý ngày tháng
if 'Ngày đặt' in df.columns:
    df['Ngày đặt'] = pd.to_datetime(df['Ngày đặt'])
elif 'Order Date' in df.columns:
    df['Ngày đặt'] = pd.to_datetime(df['Order Date'])
else:
    # Tìm cột ngày
    date_cols = [col for col in df.columns if 'ngày' in col.lower() or 'date' in col.lower()]
    if date_cols:
        df['Ngày đặt'] = pd.to_datetime(df[date_cols[0]])
    else:
        df['Ngày đặt'] = pd.date_range('2022-01-01', periods=len(df), freq='D')

# Kiểm tra và xử lý trùng lặp
print("\n🔍 KIỂM TRA DỮ LIỆU")
print("="*60)

# Kiểm tra số lượng ngày duy nhất
unique_dates = df['Ngày đặt'].nunique()
total_rows = len(df)
print(f"📊 Tổng số dòng: {total_rows}")
print(f"📊 Số ngày duy nhất: {unique_dates}")

if unique_dates < total_rows:
    print("⚠️ Phát hiện dữ liệu trùng lặp theo ngày!")
    # Gom nhóm theo ngày và tính tổng doanh số
    df = df.groupby('Ngày đặt')['Doanh số'].sum().reset_index()
    print(f"✅ Đã gom nhóm: {len(df)} dòng duy nhất theo ngày")

print(f"✅ Khoảng thời gian: {df['Ngày đặt'].min()} đến {df['Ngày đặt'].max()}")
print(f"✅ Tổng doanh số: {df['Doanh số'].sum():,.0f} VNĐ")

# Cell 3: Chuẩn bị dữ liệu chuỗi thời gian
print("\n" + "="*80)
print("CHUẨN BỊ DỮ LIỆU CHUỖI THỜI GIAN")
print("="*80)

# Đảm bảo index là datetime và sắp xếp
df = df.set_index('Ngày đặt').sort_index()
daily_sales = df['Doanh số']

# Kiểm tra tần suất
date_diff = daily_sales.index.to_series().diff().dropna()
if len(date_diff) > 0:
    most_common_freq = date_diff.value_counts().index[0]
    print(f"📊 Tần suất phổ biến nhất: {most_common_freq}")

# Doanh số theo tuần - KHÔNG cần asfreq trước
weekly_sales = daily_sales.resample('W').sum()

# Doanh số theo tháng
monthly_sales = daily_sales.resample('M').sum()

# Doanh số theo quý
quarterly_sales = daily_sales.resample('Q').sum()

print(f"📊 Dữ liệu theo ngày: {len(daily_sales)} điểm")
print(f"📊 Dữ liệu theo tuần: {len(weekly_sales)} điểm")
print(f"📊 Dữ liệu theo tháng: {len(monthly_sales)} điểm")
print(f"📊 Dữ liệu theo quý: {len(quarterly_sales)} điểm")

# Cell 4: Trực quan hóa
print("\n" + "="*80)
print("TRỰC QUAN HÓA CHUỖI THỜI GIAN")
print("="*80)

fig, axes = plt.subplots(4, 1, figsize=(16, 14))

# Daily
axes[0].plot(daily_sales.index, daily_sales.values, color='blue', alpha=0.7, linewidth=0.5)
axes[0].set_title('Doanh số hàng ngày')
axes[0].set_ylabel('Doanh số (VNĐ)')
axes[0].grid(True, alpha=0.3)

# Weekly
axes[1].plot(weekly_sales.index, weekly_sales.values, color='green', marker='o', markersize=3)
axes[1].set_title('Doanh số hàng tuần')
axes[1].set_ylabel('Doanh số (VNĐ)')
axes[1].grid(True, alpha=0.3)

# Monthly
axes[2].plot(monthly_sales.index, monthly_sales.values, color='red', marker='s', markersize=4)
axes[2].set_title('Doanh số hàng tháng')
axes[2].set_ylabel('Doanh số (VNĐ)')
axes[2].grid(True, alpha=0.3)

# Quarterly
axes[3].plot(quarterly_sales.index, quarterly_sales.values, color='purple', marker='D', markersize=5)
axes[3].set_title('Doanh số hàng quý')
axes[3].set_ylabel('Doanh số (VNĐ)')
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'time_series_overview.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 5: Phân tích thành phần (dùng dữ liệu tháng)
print("\n" + "="*80)
print("PHÂN TÍCH THÀNH PHẦN CHUỖI THỜI GIAN")
print("="*80)

# Đảm bảo đủ dữ liệu để phân tích
if len(monthly_sales) >= 24:  # Cần ít nhất 2 năm
    decomposition = seasonal_decompose(monthly_sales, model='additive', period=12)
    
    fig, axes = plt.subplots(4, 1, figsize=(16, 12))
    
    monthly_sales.plot(ax=axes[0])
    axes[0].set_title('Original - Doanh số gốc')
    axes[0].set_ylabel('Doanh số')
    axes[0].grid(True, alpha=0.3)
    
    decomposition.trend.plot(ax=axes[1])
    axes[1].set_title('Trend - Xu hướng')
    axes[1].set_ylabel('Doanh số')
    axes[1].grid(True, alpha=0.3)
    
    decomposition.seasonal.plot(ax=axes[2])
    axes[2].set_title('Seasonal - Mùa vụ')
    axes[2].set_ylabel('Doanh số')
    axes[2].grid(True, alpha=0.3)
    
    decomposition.resid.plot(ax=axes[3])
    axes[3].set_title('Residual - Phần dư')
    axes[3].set_ylabel('Doanh số')
    axes[3].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'time_series_decomposition.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("✅ Đã phân tích thành phần chuỗi thời gian")
else:
    print("⚠️ Không đủ dữ liệu để phân tích thành phần (cần ít nhất 24 tháng)")

# Cell 6: ACF và PACF (dùng dữ liệu tháng)
print("\n" + "="*80)
print("KIỂM TRA TỰ TƯƠNG QUAN")
print("="*80)

if len(monthly_sales) >= 20:
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    
    plot_acf(monthly_sales.dropna(), lags=min(20, len(monthly_sales)//2-1), ax=axes[0])
    axes[0].set_title('Autocorrelation Function (ACF)')
    
    plot_pacf(monthly_sales.dropna(), lags=min(20, len(monthly_sales)//2-1), ax=axes[1])
    axes[1].set_title('Partial Autocorrelation Function (PACF)')
    
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'acf_pacf.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("⚠️ Không đủ dữ liệu để vẽ ACF/PACF")

# Cell 7: Chia train/test (dùng dữ liệu tháng)
print("\n" + "="*80)
print("CHIA DỮ LIỆU TRAIN/TEST")
print("="*80)

# Sử dụng dữ liệu tháng
ts_data = monthly_sales.copy()
train_size = max(int(len(ts_data) * 0.8), 1)  # Đảm bảo ít nhất 1 điểm
train = ts_data[:train_size]
test = ts_data[train_size:]

print(f"📊 Tập train: {train.index[0]} đến {train.index[-1]} ({len(train)} tháng)")
print(f"📊 Tập test: {test.index[0]} đến {test.index[-1]} ({len(test)} tháng)")

if len(test) == 0:
    print("⚠️ Không đủ dữ liệu để tạo tập test!")
    # Tạo tập test giả định
    test = pd.Series([train.iloc[-1]] * 3, index=pd.date_range(train.index[-1] + pd.DateOffset(months=1), periods=3, freq='M'))
    print(f"✅ Đã tạo tập test giả định: {len(test)} tháng")

# Vẽ train/test
plt.figure(figsize=(14, 6))
plt.plot(train.index, train.values, label='Train', color='blue', linewidth=2)
plt.plot(test.index, test.values, label='Test', color='green', linewidth=2)
plt.xlabel('Thời gian')
plt.ylabel('Doanh số')
plt.title('Chia dữ liệu train/test')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'train_test_split.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 8: Hàm đánh giá
print("\n" + "="*80)
print("ĐỊNH NGHĨA HÀM ĐÁNH GIÁ")
print("="*80)

def evaluate_forecast(y_true, y_pred, model_name):
    """
    Đánh giá độ chính xác dự báo
    """
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    
    # MAPE (tránh chia cho 0)
    mask = y_true != 0
    if mask.sum() > 0:
        mape = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
    else:
        mape = np.nan
    
    # sMAPE
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    denominator = np.where(denominator == 0, 1, denominator)  # Tránh chia 0
    smape = np.mean(2 * np.abs(y_true - y_pred) / denominator) * 100
    
    print(f"\n{'='*50}")
    print(f"📊 {model_name}")
    print(f"{'='*50}")
    print(f"  MAE: {mae:,.0f} VNĐ")
    print(f"  RMSE: {rmse:,.0f} VNĐ")
    if not np.isnan(mape):
        print(f"  MAPE: {mape:.2f}%")
    print(f"  sMAPE: {smape:.2f}%")
    
    return {
        'model': model_name,
        'mae': mae,
        'rmse': rmse,
        'mape': mape if not np.isnan(mape) else 0,
        'smape': smape
    }

# Cell 9: Baseline models
print("\n" + "="*80)
print("BASELINE MODELS")
print("="*80)

results = []

# 1. Naive Forecast
naive_forecast = [train.values[-1]] * len(test)
naive_metrics = evaluate_forecast(test.values, naive_forecast, "Naive Forecast")
results.append(naive_metrics)

# 2. Seasonal Naive (nếu có đủ dữ liệu)
if len(train) >= 12:
    seasonal_naive = []
    for i in range(len(test)):
        prev_idx = i - 12
        if prev_idx >= 0 and prev_idx < len(train):
            seasonal_naive.append(train.values[prev_idx])
        else:
            seasonal_naive.append(train.values[i % len(train)])
    seasonal_metrics = evaluate_forecast(test.values, seasonal_naive, "Seasonal Naive")
    results.append(seasonal_metrics)

# 3. Moving Average
window = min(3, len(train))
ma_forecast = []
for i in range(len(test)):
    if len(train) >= window:
        ma_forecast.append(train[-window:].mean())
    else:
        ma_forecast.append(train.mean())
ma_metrics = evaluate_forecast(test.values, ma_forecast, f"MA({window})")
results.append(ma_metrics)

# Cell 10: Holt-Winters
print("\n" + "="*80)
print("HOLT-WINTERS MODEL")
print("="*80)

try:
    if len(train) >= 24:  # Cần đủ dữ liệu cho seasonal
        # Additive model
        hw_add = ExponentialSmoothing(
            train, 
            seasonal_periods=12,
            trend='add',
            seasonal='add'
        ).fit()
        
        forecast_add = hw_add.forecast(len(test))
        hw_add_metrics = evaluate_forecast(test.values, forecast_add.values, "Holt-Winters Additive")
        results.append(hw_add_metrics)
        
        # Vẽ kết quả
        plt.figure(figsize=(14, 6))
        plt.plot(train.index, train.values, label='Train', color='blue')
        plt.plot(test.index, test.values, label='Actual', color='green')
        plt.plot(test.index, forecast_add.values, label='Forecast', color='red', linestyle='--')
        if hasattr(hw_add, 'resid') and len(hw_add.resid) > 0:
            plt.fill_between(test.index, 
                            forecast_add - 1.96 * np.std(hw_add.resid),
                            forecast_add + 1.96 * np.std(hw_add.resid),
                            alpha=0.2, color='red', label='95% CI')
        plt.xlabel('Thời gian')
        plt.ylabel('Doanh số')
        plt.title('Holt-Winters Additive Forecast')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(OUTPUTS_DIR / 'holt_winters_forecast.png', dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print("⚠️ Không đủ dữ liệu cho Holt-Winters (cần ít nhất 24 tháng)")
    
except Exception as e:
    print(f"⚠️ Lỗi Holt-Winters: {e}")

# Cell 11: ARIMA
print("\n" + "="*80)
print("ARIMA MODEL")
print("="*80)

try:
    if len(train) >= 12:
        # Tìm tham số tối ưu (giới hạn để tránh quá tải)
        p = d = q = range(0, 2)
        pdq = list(itertools.product(p, d, q))
        
        best_aic = np.inf
        best_params = None
        
        print("Đang tìm tham số tối ưu...")
        
        for param in pdq:
            try:
                model = ARIMA(train, order=param)
                results_arima = model.fit()
                if results_arima.aic < best_aic:
                    best_aic = results_arima.aic
                    best_params = param
            except:
                continue
        
        if best_params:
            print(f"\n✅ Best parameters: {best_params}")
            print(f"   AIC: {best_aic:.2f}")
            
            # Huấn luyện mô hình tốt nhất
            best_arima = ARIMA(train, order=best_params)
            best_arima_fit = best_arima.fit()
            
            # Dự báo
            forecast_arima = best_arima_fit.forecast(len(test))
            arima_metrics = evaluate_forecast(test.values, forecast_arima.values, f"ARIMA{best_params}")
            results.append(arima_metrics)
            
            # Vẽ kết quả
            plt.figure(figsize=(14, 6))
            plt.plot(train.index, train.values, label='Train', color='blue')
            plt.plot(test.index, test.values, label='Actual', color='green')
            plt.plot(test.index, forecast_arima.values, label='Forecast', color='red', linestyle='--')
            plt.xlabel('Thời gian')
            plt.ylabel('Doanh số')
            plt.title('ARIMA Forecast')
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.savefig(OUTPUTS_DIR / 'arima_forecast.png', dpi=300, bbox_inches='tight')
            plt.show()
        else:
            print("⚠️ Không tìm được tham số ARIMA phù hợp")
    else:
        print("⚠️ Không đủ dữ liệu cho ARIMA (cần ít nhất 12 tháng)")
    
except Exception as e:
    print(f"⚠️ Lỗi ARIMA: {e}")

# Cell 12: So sánh các mô hình
print("\n" + "="*80)
print("SO SÁNH CÁC MÔ HÌNH")
print("="*80)

if results:
    # Tạo bảng so sánh
    comparison_df = pd.DataFrame(results)
    comparison_df = comparison_df.sort_values('smape')
    
    print("\n📊 BẢNG SO SÁNH:")
    print(comparison_df.to_string(index=False))
    
    # Vẽ biểu đồ so sánh
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # MAE
    models = comparison_df['model'].values
    mae_values = comparison_df['mae'].values
    colors = ['gold' if x == min(mae_values) else 'skyblue' for x in mae_values]
    bars1 = axes[0].barh(models, mae_values, color=colors)
    axes[0].set_xlabel('MAE (VNĐ)')
    axes[0].set_title('So sánh MAE')
    for bar, val in zip(bars1, mae_values):
        axes[0].text(bar.get_width() + 50000, bar.get_y() + bar.get_height()/2, 
                    f'{val:,.0f}', va='center')
    
    # RMSE
    rmse_values = comparison_df['rmse'].values
    colors = ['gold' if x == min(rmse_values) else 'skyblue' for x in rmse_values]
    bars2 = axes[1].barh(models, rmse_values, color=colors)
    axes[1].set_xlabel('RMSE (VNĐ)')
    axes[1].set_title('So sánh RMSE')
    for bar, val in zip(bars2, rmse_values):
        axes[1].text(bar.get_width() + 50000, bar.get_y() + bar.get_height()/2, 
                    f'{val:,.0f}', va='center')
    
    # sMAPE
    smape_values = comparison_df['smape'].values
    colors = ['gold' if x == min(smape_values) else 'skyblue' for x in smape_values]
    bars3 = axes[2].barh(models, smape_values, color=colors)
    axes[2].set_xlabel('sMAPE (%)')
    axes[2].set_title('So sánh sMAPE')
    for bar, val in zip(bars3, smape_values):
        axes[2].text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2, 
                    f'{val:.2f}%', va='center')
    
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'forecast_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Cell 13: Dự báo tương lai
    print("\n" + "="*80)
    print("DỰ BÁO 12 THÁNG TIẾP THEO")
    print("="*80)
    
    # Chọn mô hình tốt nhất
    best_idx = comparison_df['smape'].idxmin()
    best_model_name = comparison_df.loc[best_idx, 'model']
    print(f"🏆 Mô hình tốt nhất: {best_model_name}")
    print(f"   sMAPE: {comparison_df.loc[best_idx, 'smape']:.2f}%")
    
    # Dự báo 12 tháng
    if 'ARIMA' in best_model_name and 'best_arima_fit' in locals():
        forecast_future = best_arima_fit.forecast(12)
        forecast_index = pd.date_range(start=ts_data.index[-1] + pd.DateOffset(months=1), 
                                       periods=12, freq='M')
        
        print("\n📊 Dự báo 12 tháng tới:")
        for date, value in zip(forecast_index, forecast_future):
            print(f"   {date.strftime('%Y-%m')}: {value:,.0f} VNĐ")
        
        # Vẽ dự báo
        plt.figure(figsize=(14, 6))
        plt.plot(ts_data.index, ts_data.values, label='Lịch sử', color='blue')
        plt.plot(forecast_index, forecast_future, label='Dự báo', color='red', linestyle='--', marker='o')
        plt.xlabel('Thời gian')
        plt.ylabel('Doanh số')
        plt.title('Dự báo 12 tháng tiếp theo')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(OUTPUTS_DIR / 'future_forecast.png', dpi=300, bbox_inches='tight')
        plt.show()
    else:
        # Dùng naive forecast
        last_value = train.values[-1]
        forecast_future = [last_value] * 12
        forecast_index = pd.date_range(start=ts_data.index[-1] + pd.DateOffset(months=1), 
                                       periods=12, freq='M')
        print("\n📊 Dự báo 12 tháng tới (Naive):")
        for date, value in zip(forecast_index, forecast_future):
            print(f"   {date.strftime('%Y-%m')}: {value:,.0f} VNĐ")
else:
    print("⚠️ Không có kết quả để so sánh!")

# Cell 14: Phân tích theo mùa
print("\n" + "="*80)
print("PHÂN TÍCH THEO MÙA")
print("="*80)

# Phân tích theo tháng
if len(daily_sales) >= 365:
    monthly_pattern = df.resample('M')['Doanh số'].mean()
    monthly_pattern.index = monthly_pattern.index.month
    monthly_pattern = monthly_pattern.groupby(level=0).mean()
    monthly_pattern.index = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9', 'T10', 'T11', 'T12']
    
    print("📊 Doanh số trung bình theo tháng:")
    print(monthly_pattern)
    
    # Vẽ biểu đồ
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Bar chart
    axes[0].bar(range(1, 13), monthly_pattern.values, color='skyblue', edgecolor='black')
    axes[0].set_xlabel('Tháng')
    axes[0].set_ylabel('Doanh số trung bình')
    axes[0].set_title('Doanh số trung bình theo tháng')
    axes[0].set_xticks(range(1, 13))
    
    # Heatmap
    pivot_table = df.resample('M')['Doanh số'].sum().to_frame()
    pivot_table['year'] = pivot_table.index.year
    pivot_table['month'] = pivot_table.index.month
    pivot_table = pivot_table.pivot_table(values='Doanh số', index='year', columns='month', aggfunc='mean')
    
    if not pivot_table.empty:
        sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlOrRd', ax=axes[1])
        axes[1].set_xlabel('Tháng')
        axes[1].set_ylabel('Năm')
        axes[1].set_title('Heatmap doanh số')
    
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'seasonal_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print("⚠️ Không đủ dữ liệu để phân tích theo mùa")

# Cell 15: Lưu kết quả
print("\n" + "="*80)
print("LƯU KẾT QUẢ")
print("="*80)

if results:
    # Lưu so sánh
    comparison_df.to_csv(OUTPUTS_DIR / 'forecast_comparison.csv', index=False, encoding='utf-8-sig')
    
    # Lưu dữ liệu đã xử lý
    monthly_sales.to_csv(OUTPUTS_DIR / 'monthly_sales.csv', encoding='utf-8-sig')
    weekly_sales.to_csv(OUTPUTS_DIR / 'weekly_sales.csv', encoding='utf-8-sig')
    
    print(f"✅ Đã lưu tại: {OUTPUTS_DIR}")
    print(f"   - forecast_comparison.csv: So sánh các mô hình")
    print(f"   - monthly_sales.csv: Doanh số theo tháng")
    print(f"   - weekly_sales.csv: Doanh số theo tuần")

# Cell 16: Kết luận
print("\n" + "="*80)
print("KẾT LUẬN")
print("="*80)

if results:
    best_model = comparison_df.loc[best_idx]
    print(f"""
1. TỔNG QUAN:
   - Dữ liệu: {len(ts_data)} tháng
   - Thời gian: {ts_data.index[0]} đến {ts_data.index[-1]}
   - Tần suất: Tháng

2. KẾT QUẢ TỐT NHẤT:
   - Mô hình: {best_model['model']}
   - sMAPE: {best_model['smape']:.2f}%
   - MAE: {best_model['mae']:,.0f} VNĐ
   - RMSE: {best_model['rmse']:,.0f} VNĐ

3. KHUYẾN NGHỊ:
   - Tăng inventory cho các tháng cao điểm
   - Chuẩn bị marketing trước mùa cao điểm
   - Theo dõi sát biến động thị trường
""")
else:
    print("⚠️ Không có kết quả phân tích!")

print("\n" + "="*80)
print("✅ HOÀN THÀNH PHÂN TÍCH CHUỖI THỜI GIAN!")
print("="*80)
```
