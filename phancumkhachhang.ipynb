```python
# Cell 1: Import thư viện
print("="*80)
print("PHÂN CỤM KHÁCH HÀNG - RFM ANALYSIS")
print("Sử dụng bộ dữ liệu thực tế: Sales and Inventory Snapshot Data")
print("="*80)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import warnings
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Thiết lập hiển thị
plt.rcParams['figure.figsize'] = (14, 8)
sns.set_style('whitegrid')
pd.set_option('display.max_columns', None)

print("✅ Đã import thư viện thành công!")

# Cell 2: Đường dẫn và đọc dữ liệu
print("\n" + "="*80)
print("ĐỌC DỮ LIỆU")
print("="*80)

# Xác định đường dẫn
PROJECT_DIR = Path.cwd() / 'vietnam_retail_analysis'
DATA_DIR = PROJECT_DIR / 'data' / 'processed'
OUTPUTS_DIR = PROJECT_DIR / 'outputs'

print(f"📁 Thư mục dữ liệu: {DATA_DIR}")
print(f"📁 Thư mục kết quả: {OUTPUTS_DIR}")

# Tìm file dữ liệu đã xử lý
sales_file = DATA_DIR / 'combined_sales_data.csv'
if not sales_file.exists():
    # Thử tìm trong thư mục data gốc
    sales_file = PROJECT_DIR / 'data' / 'raw' / 'combined_sales_data.csv'
    if not sales_file.exists():
        print("⚠️ Không tìm thấy file dữ liệu đã xử lý!")
        print("Đang tạo dữ liệu mẫu...")
        
        # Tạo dữ liệu mẫu
        np.random.seed(42)
        n_customers = 500
        n_transactions = 5000
        
        customers = [f'KH-{i:05d}' for i in range(1, n_customers + 1)]
        dates = pd.date_range('2022-01-01', '2023-12-31', freq='D')
        
        data = []
        for i in range(n_transactions):
            customer = np.random.choice(customers)
            order_date = np.random.choice(dates)
            data.append({
                'Mã khách hàng': customer,
                'Ngày đặt': order_date,
                'Mã đơn hàng': f'DH-{i+1000:06d}',
                'Doanh số': np.random.randint(100000, 5000000),
                'Số lượng': np.random.randint(1, 10)
            })
        
        df = pd.DataFrame(data)
        print(f"✅ Đã tạo dữ liệu mẫu: {len(df)} giao dịch")
    else:
        df = pd.read_csv(sales_file, encoding='utf-8-sig')
        print(f"✅ Đã đọc file: {sales_file}")
else:
    df = pd.read_csv(sales_file, encoding='utf-8-sig')
    print(f"✅ Đã đọc file: {sales_file}")

print(f"✅ Kích thước dữ liệu: {df.shape}")
print(f"✅ Các cột: {df.columns.tolist()}")

# Cell 3: Chuẩn bị dữ liệu
print("\n" + "="*80)
print("CHUẨN BỊ DỮ LIỆU")
print("="*80)

# Chuyển đổi kiểu dữ liệu
if 'Ngày đặt' in df.columns:
    df['Ngày đặt'] = pd.to_datetime(df['Ngày đặt'])
elif 'Order Date' in df.columns:
    df['Ngày đặt'] = pd.to_datetime(df['Order Date'])
else:
    # Tìm cột ngày
    date_cols = [col for col in df.columns if 'ngày' in col.lower() or 'date' in col.lower()]
    if date_cols:
        df['Ngày đặt'] = pd.to_datetime(df[date_cols[0]])
    else:
        # Tạo cột ngày giả
        df['Ngày đặt'] = pd.date_range('2022-01-01', periods=len(df), freq='D')

# Xác định ngày phân tích
snapshot_date = df['Ngày đặt'].max() + pd.Timedelta(days=1)
print(f"📅 Ngày phân tích: {snapshot_date}")

# Xác định cột mã khách hàng
if 'Mã khách hàng' in df.columns:
    customer_col = 'Mã khách hàng'
elif 'Customer ID' in df.columns:
    customer_col = 'Customer ID'
else:
    # Tìm cột có chứa 'khách' hoặc 'customer'
    customer_cols = [col for col in df.columns if 'khách' in col.lower() or 'customer' in col.lower()]
    if customer_cols:
        customer_col = customer_cols[0]
    else:
        customer_col = df.columns[0]
        print(f"⚠️ Sử dụng cột {customer_col} làm mã khách hàng")

# Xác định cột đơn hàng
if 'Mã đơn hàng' in df.columns:
    order_col = 'Mã đơn hàng'
elif 'Order ID' in df.columns:
    order_col = 'Order ID'
else:
    # Tìm cột có chứa 'đơn' hoặc 'order'
    order_cols = [col for col in df.columns if 'đơn' in col.lower() or 'order' in col.lower()]
    if order_cols:
        order_col = order_cols[0]
    else:
        order_col = df.columns[1] if len(df.columns) > 1 else df.columns[0]
        print(f"⚠️ Sử dụng cột {order_col} làm mã đơn hàng")

# Xác định cột doanh số
if 'Doanh số' in df.columns:
    sales_col = 'Doanh số'
elif 'Sales' in df.columns:
    sales_col = 'Sales'
else:
    # Tìm cột có chứa 'doanh' hoặc 'sales' hoặc 'revenue'
    sales_cols = [col for col in df.columns if 'doanh' in col.lower() or 'sales' in col.lower() or 'revenue' in col.lower()]
    if sales_cols:
        sales_col = sales_cols[0]
    else:
        # Nếu không tìm thấy, chọn cột số đầu tiên
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        sales_col = numeric_cols[0] if numeric_cols else df.columns[2]
        print(f"⚠️ Sử dụng cột {sales_col} làm doanh số")

print(f"\n✅ Các cột được xác định:")
print(f"   - Khách hàng: {customer_col}")
print(f"   - Đơn hàng: {order_col}")
print(f"   - Doanh số: {sales_col}")

# Cell 4: Tính toán RFM
print("\n" + "="*80)
print("TÍNH TOÁN RFM")
print("="*80)

# Tính RFM cơ bản
rfm = df.groupby(customer_col).agg({
    'Ngày đặt': lambda x: (snapshot_date - x.max()).days,  # Recency
    order_col: 'nunique',  # Frequency
    sales_col: 'sum'  # Monetary
}).rename(columns={
    'Ngày đặt': 'Recency',
    order_col: 'Frequency',
    sales_col: 'Monetary'
})

print(f"✅ Đã tính RFM cho {len(rfm)} khách hàng")
print("\n📊 5 dòng đầu:")
print(rfm.head())

# Thêm các đặc trưng khác (linh hoạt theo dữ liệu có sẵn)
agg_dict = {}

# Thêm cột số lượng nếu có
if 'Số lượng' in df.columns:
    agg_dict['Số lượng'] = 'sum'
elif 'Quantity' in df.columns:
    agg_dict['Quantity'] = 'sum'

# Thêm cột lợi nhuận nếu có
if 'Lợi nhuận' in df.columns:
    agg_dict['Lợi nhuận'] = 'sum'
elif 'Profit' in df.columns:
    agg_dict['Profit'] = 'sum'

# Thêm các cột số khác có thể hữu ích
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
other_cols = [col for col in numeric_cols if col not in [sales_col, 'Số lượng', 'Quantity', 'Lợi nhuận', 'Profit']]
for col in other_cols[:3]:  # Chỉ lấy tối đa 3 cột số khác
    agg_dict[col] = 'mean'

if agg_dict:
    extra_features = df.groupby(customer_col).agg(agg_dict)
    
    # Đổi tên cột
    rename_dict = {}
    for col in agg_dict.keys():
        if 'lượng' in col.lower() or 'quantity' in col.lower():
            rename_dict[col] = 'Total_Quantity'
        elif 'lợi' in col.lower() or 'profit' in col.lower():
            rename_dict[col] = 'Total_Profit'
        else:
            rename_dict[col] = f'Avg_{col}'
    
    extra_features = extra_features.rename(columns=rename_dict)
    
    # Kết hợp
    rfm_full = rfm.join(extra_features, how='left').fillna(0)
    print(f"\n✅ Đã thêm {len(extra_features.columns)} đặc trưng bổ sung")
else:
    rfm_full = rfm.copy()
    print("\n⚠️ Không có đặc trưng bổ sung")

print("\n📊 RFM đầy đủ:")
print(rfm_full.head())

# Cell 5: Thống kê mô tả
print("\n" + "="*80)
print("THỐNG KÊ MÔ TẢ RFM")
print("="*80)

print(rfm_full.describe())

# Vẽ phân bố
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
features = rfm_full.columns[:6]  # Lấy 6 cột đầu

for i, feature in enumerate(features):
    row = i // 3
    col = i % 3
    if feature in rfm_full.columns:
        axes[row, col].hist(rfm_full[feature], bins=30, edgecolor='black', alpha=0.7, color='skyblue')
        axes[row, col].set_title(f'Phân bố {feature}')
        axes[row, col].set_xlabel(feature)
        axes[row, col].set_ylabel('Số lượng')
        axes[row, col].axvline(rfm_full[feature].mean(), color='red', linestyle='--', label='Mean')
        axes[row, col].axvline(rfm_full[feature].median(), color='green', linestyle='--', label='Median')
        axes[row, col].legend()

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'rfm_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 6: Chuẩn hóa dữ liệu
print("\n" + "="*80)
print("CHUẨN HÓA DỮ LIỆU")
print("="*80)

# Chọn features cho phân cụm
cluster_features = ['Recency', 'Frequency', 'Monetary']
available_features = [f for f in cluster_features if f in rfm_full.columns]

# Thêm các features bổ sung nếu có
for col in rfm_full.columns:
    if col not in available_features and col not in ['Recency', 'Frequency', 'Monetary']:
        if len(available_features) < 5:  # Giới hạn số features
            available_features.append(col)

print(f"Features sử dụng: {available_features}")

X = rfm_full[available_features].copy()

# Xử lý giá trị vô cùng
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median())

# Chuẩn hóa
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"✅ Đã chuẩn hóa {X_scaled.shape[1]} features cho {X_scaled.shape[0]} khách hàng")

# Cell 7: Tìm số cụm tối ưu
print("\n" + "="*80)
print("TÌM SỐ CỤM TỐI ƯU")
print("="*80)

inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    sil_score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(sil_score)
    print(f"  - k={k}: Inertia={kmeans.inertia_:.0f}, Silhouette={sil_score:.4f}")

# Vẽ biểu đồ Elbow
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(K_range, inertias, 'bo-')
ax1.set_xlabel('Số cụm (k)')
ax1.set_ylabel('Inertia')
ax1.set_title('Phương pháp Elbow')
ax1.grid(True)
# Đánh dấu elbow point
elbow_k = 4
if elbow_k <= len(inertias):
    ax1.plot(elbow_k, inertias[elbow_k-2], 'ro', markersize=10, label=f'Elbow point (k={elbow_k})')
ax1.legend()

ax2.plot(K_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Số cụm (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score')
ax2.grid(True)
# Đánh dấu max silhouette
max_k = K_range[np.argmax(silhouette_scores)]
ax2.plot(max_k, max(silhouette_scores), 'go', markersize=10, label=f'Max at k={max_k}')
ax2.legend()

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'optimal_clusters.png', dpi=300, bbox_inches='tight')
plt.show()

# Chọn số cụm
optimal_k = max_k
print(f"\n✅ Chọn số cụm tối ưu: k={optimal_k}")

# Cell 8: Thực hiện phân cụm
print("\n" + "="*80)
print(f"PHÂN CỤM VỚI K={optimal_k}")
print("="*80)

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
rfm_full['Cluster'] = kmeans.fit_predict(X_scaled)

# Phân bố khách hàng
cluster_dist = rfm_full['Cluster'].value_counts().sort_index()
print("📊 Phân bố khách hàng theo cụm:")
for cluster, count in cluster_dist.items():
    print(f"  - Cụm {cluster}: {count} khách hàng ({count/len(rfm_full)*100:.1f}%)")

# Cell 9: Phân tích đặc điểm từng cụm
print("\n" + "="*80)
print("PHÂN TÍCH ĐẶC ĐIỂM CỤM")
print("="*80)

# Tính trung bình các đặc trưng theo cụm
cluster_profile = rfm_full.groupby('Cluster').mean().round(2)

print("\n📊 Hồ sơ cụm (giá trị trung bình):")
print(cluster_profile)

# Thêm số lượng và tỷ lệ
cluster_profile['Số lượng'] = cluster_dist.values
cluster_profile['Tỷ lệ %'] = (cluster_dist.values / len(rfm_full) * 100).round(2)

print("\n📊 Quy mô cụm:")
print(cluster_profile[['Số lượng', 'Tỷ lệ %']])

# Cell 10: Đặt tên và mô tả cụm
print("\n" + "="*80)
print("ĐẶT TÊN VÀ MÔ TẢ CỤM")
print("="*80)

cluster_names = {}
cluster_descriptions = {}

for cluster in range(optimal_k):
    cluster_data = rfm_full[rfm_full['Cluster'] == cluster]
    
    # Phân tích đặc điểm
    avg_recency = cluster_data['Recency'].mean()
    avg_monetary = cluster_data['Monetary'].mean()
    avg_frequency = cluster_data['Frequency'].mean()
    
    # Đặt tên dựa trên đặc điểm
    if avg_recency < rfm_full['Recency'].median():
        if avg_monetary > rfm_full['Monetary'].median():
            name = "💎 VIP - Khách hàng thân thiết"
            desc = "Chi tiêu cao, mua gần đây, tần suất cao"
        else:
            name = "🌟 Tiềm năng - Mới mua"
            desc = "Mua gần đây nhưng chi tiêu chưa cao"
    else:
        if avg_monetary > rfm_full['Monetary'].median():
            name = "📦 Cũ - Chi tiêu cao"
            desc = "Đã lâu không mua nhưng từng chi tiêu lớn"
        else:
            name = "⚠️ Rủi ro - Chi tiêu thấp"
            desc = "Đã lâu không mua, chi tiêu thấp"
    
    cluster_names[cluster] = name
    cluster_descriptions[cluster] = desc
    
    print(f"\n📌 Cụm {cluster}: {name}")
    print(f"   {desc}")
    print(f"   Số lượng: {len(cluster_data)} khách hàng")
    print(f"   Recency TB: {avg_recency:.0f} ngày")
    print(f"   Frequency TB: {avg_frequency:.1f} đơn")
    print(f"   Monetary TB: {avg_monetary:,.0f} VNĐ")

# Thêm tên cụm
rfm_full['Cluster_Name'] = rfm_full['Cluster'].map(cluster_names)

# Cell 11: Trực quan hóa PCA
print("\n" + "="*80)
print("TRỰC QUAN HÓA PCA 2D")
print("="*80)

# Giảm chiều với PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Vẽ biểu đồ
plt.figure(figsize=(12, 8))
colors = plt.cm.viridis(np.linspace(0, 1, optimal_k))

for cluster in range(optimal_k):
    mask = rfm_full['Cluster'] == cluster
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1], 
               c=[colors[cluster]], label=cluster_names[cluster], 
               alpha=0.6, s=50)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%})')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%})')
plt.title('Phân bố khách hàng theo cụm (PCA 2D)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'pca_clusters.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 12: Biểu đồ boxplot so sánh
print("\n" + "="*80)
print("SO SÁNH CÁC CỤM")
print("="*80)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
plot_features = ['Recency', 'Frequency', 'Monetary'] + [col for col in rfm_full.columns if col not in ['Recency', 'Frequency', 'Monetary', 'Cluster', 'Cluster_Name']][:3]

for i, feature in enumerate(plot_features[:6]):  # Tối đa 6 features
    if feature in rfm_full.columns:
        row = i // 3
        col = i % 3
        
        data_to_plot = [rfm_full[rfm_full['Cluster'] == cluster][feature] for cluster in range(optimal_k)]
        bp = axes[row, col].boxplot(data_to_plot, patch_artist=True)
        
        # Tô màu
        for patch, color in zip(bp['boxes'], colors[:len(data_to_plot)]):
            patch.set_facecolor(color)
        
        axes[row, col].set_xlabel('Cụm')
        axes[row, col].set_ylabel(feature)
        axes[row, col].set_title(f'So sánh {feature}')
        axes[row, col].set_xticklabels([f'Cụm {i}' for i in range(optimal_k)])
        axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(OUTPUTS_DIR / 'cluster_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 13: Chiến lược marketing
print("\n" + "="*80)
print("CHIẾN LƯỢC MARKETING CHO TỪNG CỤM")
print("="*80)

marketing_strategies = {
    "💎 VIP - Khách hàng thân thiết": [
        "✓ Chương trình VIP với ưu đãi đặc biệt",
        "✓ Ưu tiên giới thiệu sản phẩm mới",
        "✓ Tặng quà tri ân vào dịp đặc biệt",
        "✓ Tích điểm thưởng cao hơn",
        "✓ Mời tham gia sự kiện riêng"
    ],
    "🌟 Tiềm năng - Mới mua": [
        "✓ Khuyến khích mua hàng lần 2",
        "✓ Gửi email chào mừng và ưu đãi",
        "✓ Thu thập feedback về trải nghiệm",
        "✓ Tư vấn sản phẩm phù hợp",
        "✓ Giảm giá cho đơn hàng tiếp theo"
    ],
    "📦 Cũ - Chi tiêu cao": [
        "✓ Chiến dịch tái kích hoạt",
        "✓ Gửi ưu đãi đặc biệt qua email/SMS",
        "✓ Khảo sát lý do không mua gần đây",
        "✓ Giới thiệu sản phẩm mới tương tự",
        "✓ Ưu đãi theo mùa"
    ],
    "⚠️ Rủi ro - Chi tiêu thấp": [
        "✓ Gửi ưu đãi giảm giá mạnh",
        "✓ Tập trung vào sản phẩm giá rẻ",
        "✓ Khuyến khích mua theo combo",
        "✓ Email nhắc nhở với nội dung hấp dẫn",
        "✓ Tặng voucher khi quay lại"
    ]
}

for cluster in range(optimal_k):
    name = cluster_names[cluster]
    print(f"\n📌 CỤM {cluster}: {name}")
    print(f"   Số lượng: {cluster_dist[cluster]} khách hàng ({cluster_profile.loc[cluster, 'Tỷ lệ %']:.1f}%)")
    print("\n   Chiến lược:")
    
    # Tìm chiến lược phù hợp
    strategy_key = None
    for key in marketing_strategies.keys():
        if key.split(' - ')[0] in name:
            strategy_key = key
            break
    
    if strategy_key:
        for strategy in marketing_strategies[strategy_key]:
            print(f"   {strategy}")
    else:
        print("   ✓ Chưa có chiến lược cụ thể cho cụm này")

# Cell 14: Lưu kết quả
print("\n" + "="*80)
print("LƯU KẾT QUẢ")
print("="*80)

# Lưu dữ liệu đã gán nhãn
rfm_full.to_csv(OUTPUTS_DIR / 'customer_segments.csv', encoding='utf-8-sig')

# Lưu hồ sơ cụm
cluster_profile.to_csv(OUTPUTS_DIR / 'cluster_profile.csv', encoding='utf-8-sig')

# Lưu thông tin chi tiết
cluster_summary = pd.DataFrame({
    'Cụm': list(cluster_names.keys()),
    'Tên cụm': list(cluster_names.values()),
    'Mô tả': list(cluster_descriptions.values()),
    'Số lượng': cluster_dist.values,
    'Tỷ lệ %': (cluster_dist.values / len(rfm_full) * 100).round(2)
})
cluster_summary.to_csv(OUTPUTS_DIR / 'cluster_summary.csv', index=False, encoding='utf-8-sig')

print(f"✅ Đã lưu các file tại: {OUTPUTS_DIR}")
print(f"   - customer_segments.csv: Phân cụm từng khách hàng")
print(f"   - cluster_profile.csv: Hồ sơ chi tiết từng cụm")
print(f"   - cluster_summary.csv: Tổng hợp thông tin các cụm")

# Cell 15: Kết luận
print("\n" + "="*80)
print("KẾT LUẬN PHÂN CỤM KHÁCH HÀNG")
print("="*80)

print(f"""
1. TỔNG QUAN:
   - Số cụm: {optimal_k}
   - Tổng số khách hàng: {len(rfm_full)}
   - Phương pháp: K-Means với RFM

2. ĐẶC ĐIỂM CÁC CỤM:
""")

for cluster in range(optimal_k):
    print(f"""
   Cụm {cluster}: {cluster_names[cluster]}
   - Số lượng: {cluster_dist[cluster]} ({cluster_profile.loc[cluster, 'Tỷ lệ %']:.1f}%)
   - Recency TB: {cluster_profile.loc[cluster, 'Recency']:.0f} ngày
   - Frequency TB: {cluster_profile.loc[cluster, 'Frequency']:.1f} đơn
   - Monetary TB: {cluster_profile.loc[cluster, 'Monetary']:,.0f} VNĐ
   """)

print("""
3. GIÁ TRỊ KINH DOANH:
   - Cá nhân hóa marketing theo từng cụm
   - Tối ưu ngân sách quảng cáo
   - Tăng tỷ lệ giữ chân khách hàng
   - Phát triển sản phẩm phù hợp từng phân khúc
""")

print("\n" + "="*80)
print("✅ HOÀN THÀNH PHÂN CỤM KHÁCH HÀNG!")
print("="*80)
```
